### Basic Imports 

```python
import pandas as pd
import numpy as np
```

```python
cat_df = pd.read_csv("../Datasets/whiskers_and_purrs.csv")
cat_df.head()
cat_df.shape
```

```python
import matplotlib.pyplot as plt
import pandas as pd

class SleepDataAnalyzer:
    def __init__(self, sleep_data):
        self.sleep_data = sleep_data
        self.mean_hours = self.sleep_data.mean()
        self.median_hours = self.sleep_data.median()
        self.mode_hours = self.sleep_data.mode()[0]

    def plot_histogram(self):
        # Prepare the histogram
        plt.figure(figsize=(10, 6))
        plt.hist(self.sleep_data, bins=10, alpha=0.7, color='skyblue', edgecolor='black')

        # Plotting mean, median, and mode lines
        plt.axvline(self.mean_hours, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {self.mean_hours:.2f}')
        plt.axvline(self.median_hours, color='green', linestyle='dashed', linewidth=2, label=f'Median: {self.median_hours:.2f}')
        plt.axvline(self.mode_hours, color='purple', linestyle='dashed', linewidth=2, label=f'Mode: {self.mode_hours:.2f}')

        # Adding labels and title
        plt.title('Histogram of Hours of Sleep for Cats')
        plt.xlabel('Hours of Sleep (per day)')
        plt.ylabel('Frequency')
        plt.legend()
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
```

### With Missing values

```python
sleep_data = cat_df['Hours of Sleep (per day)']
analyzer = SleepDataAnalyzer(sleep_data)
analyzer.plot_histogram()
```

### Putting NA as mean  

```python
sleep_data = cat_df['Hours of Sleep (per day)'].fillna(sleep_data.mean(numeric_only=True))
analyzer = SleepDataAnalyzer(sleep_data)
analyzer.plot_histogram()
```

### Putting NA as median 

```python
sleep_data = cat_df['Hours of Sleep (per day)'].fillna(sleep_data.median(numeric_only=True))
analyzer = SleepDataAnalyzer(sleep_data)
analyzer.plot_histogram()
```

### Putting NA as mode

```python
sleep_data = cat_df['Hours of Sleep (per day)'].fillna(sleep_data.mode()[0])
analyzer = SleepDataAnalyzer(sleep_data)
analyzer.plot_histogram()
```
